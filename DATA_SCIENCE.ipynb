```python
# 04 Stepwise selection
import statsmodels.api as sm
import pandas as pd
import numpy as np

# loading the data
data = pd.read_csv("/content/COLOR COMBINATION DATASET FINAL.csv")

# Define the dependent variable (y) and independent variables (X)

y = data['Serial Number']
# Select relevant numerical columns and drop the target variable
X = data.select_dtypes(include=np.number).drop(columns=['Serial Number'])

# Add a constant column for the intercept
X_with_const = sm.add_constant(X)

# Initialize the list of selected features
selected_features = list(X.columns) # Start with all features

# Initialize the list of p-values
p_values = []

# Perform stepwise selection
while True:
    # Fit the model with the current set of features
    model = sm.OLS(y, X_with_const[selected_features + ['const']]).fit() # Correctly index X_with_const

    # Get the p-values of the current features
    p_values = model.pvalues[1:] # Exclude the constant's p-value

    # Find the feature with the highest p-value
    if p_values.empty: # Handle the case when all features are removed
        break
    max_p_value = max(p_values)
    max_p_feature = p_values.idxmax()

    # If the highest p-value is less than the significance level (e.g., 0.05), stop
    if max_p_value < 0.05:
        break

    # Remove the feature with the highest p-value
    selected_features.remove(max_p_feature)

# Display the summary of the final model
print(model.summary())
```

```python
# 01 Principal Component Analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# loading the data
data = pd.read_csv("/content/COLOR COMBINATION DATASET FINAL.csv")

# Standardize the data
features = ['Green', 'Red', 'Blue', 'Yellow', 'Orange', 'Pink', 'Violet', 'Cyan']
scaled_data = StandardScaler().fit_transform(data[features])

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Plot the principal components
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.title('PCA Result')
plt.xlabel('Combination')
plt.ylabel('BACKGROUND')
plt.show()

# Calculate the individual score of each PRINCIPAL COMPONENT
individual_scores = pca.explained_variance_ratio_
# Normalize scores (this step might be optional depending on your interpretation)
individual_scores = individual_scores / individual_scores.sum()

# Create a DataFrame with PRINCIPAL COMPONENT names and individual scores
pc_names = ['BACKGROUND', 'Combination']  # Names of principal components
feature_scores = pd.DataFrame({'Feature': pc_names, 'Score': individual_scores})

# Calculate the overall score of the PCA algorithm
overall_score = np.mean(individual_scores)

# Print the individual scores and overall score
print("\nIndividual scores of each feature:")
print(feature_scores)

print("\nOverall score of the PCA algorithm:")
print(f"Overall Score: {overall_score}")

# Plot the individual scores
plt.figure(figsize=(10, 6))
plt.bar(feature_scores['Feature'], feature_scores['Score'])
plt.title("Individual Scores of Each Feature")
plt.xlabel("Feature")
plt.ylabel("Score")
plt.xticks(rotation=90)
plt.show()
```

```python
# 02 Chi-square test

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder

# loading the data
data = pd.read_csv("/content/COLOR COMBINATION DATASET FINAL.csv")

# Converting categorical variables to numeric if necessary
# Remove the extra '\n' from column names
X = data.drop(columns=['Output Color', 'Serial Number', 'Color Code', 'BACKGROUND'])  # Drop non-numerical columns

le = LabelEncoder()
# Also remove '\n' here for consistency
y = le.fit_transform(data['Output Color'])

# Chi-square test
chi_scores = chi2(X, y)

# Creating a DataFrame with chi-square scores
chi2_df = pd.DataFrame({'Feature': X.columns, 'Chi2': chi_scores[0], 'P-value': chi_scores[1]})
chi2_df = chi2_df.sort_values(by='Chi2', ascending=False)

print("Top features based on Chi-Square test:")
print(chi2_df.head(10))

# Calculate the individual score of each feature
total_chi2 = chi2_df['Chi2'].sum()
chi2_df['Score'] = chi2_df['Chi2'] / total_chi2

# Calculate the overall score of the Chi-Square algorithm
overall_score = chi2_df['Score'].mean()

print("\nIndividual scores of each feature:")
print(chi2_df)

print("\nOverall score of the Chi-Square algorithm:")
print("Overall Score:", overall_score)

# Plot the scores of each feature
plt.figure(figsize=(10, 6))
sns.barplot(x='Feature', y='Score', data=chi2_df)
plt.title("Individual Scores of Each Feature")
plt.xlabel("Feature")
plt.ylabel("Score")
plt.xticks(rotation=90)
plt.show()
```

```python
# 04 Stepwise selection

import statsmodels.api as sm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# loading the data
data = pd.read_csv("/content/COLOR COMBINATION DATASET FINAL.csv")

# Define the dependent variable (y) and independent variables (X)

y = data['Serial Number']
# Select relevant numerical columns and drop the target variable
X = data.select_dtypes(include=np.number).drop(columns=['Serial Number'])

# Add a constant column for the intercept
X_with_const = sm.add_constant(X)

# Initialize the list of selected features
selected_features = list(X.columns) # Start with all features

# Initialize the list of p-values
p_values = []

# Perform stepwise selection
while True:
    # Fit the model with the current set of features
    model = sm.OLS(y, X_with_const[selected_features + ['const']]).fit() # Correctly index X_with_const

    # Get the p-values of the current features
    p_values = model.pvalues[1:] # Exclude the constant's p-value

    # Find the feature with the highest p-value
    if p_values.empty: # Handle the case when all features are removed
        break
    max_p_value = max(p_values)
    max_p_feature = p_values.idxmax()

    # If the highest p-value is less than the significance level (e.g., 0.05), stop
    if max_p_value < 0.05:
        break

    # Remove the feature with the highest p-value
    selected_features.remove(max_p_feature)

# Display the summary of the final model
print(model.summary())

# Calculate the individual score of each feature
individual_scores = model.pvalues[1:] # Exclude the constant's p-value
individual_scores = 1 - individual_scores # Convert p-values to scores

# Create a DataFrame with feature names and individual scores
feature_scores = pd.DataFrame({'Feature': individual_scores.index, 'Score': individual_scores.values})

# Calculating the overall score of the Stepwise selection algorithm
overall_score = np.mean(individual_scores)

# Print the individual scores and overall score
print("\nIndividual scores of each feature:")
print(feature_scores)

print("\nOverall score of the Stepwise selection algorithm:")
print(f"Overall Score: {overall_score}")

# Plot the individual scores
plt.figure(figsize=(10, 6))
plt.bar(feature_scores['Feature'], feature_scores['Score'])
plt.title("Individual Scores of Each Feature")
plt.xlabel("Feature")
plt.ylabel("Score")
plt.xticks(rotation=90)
plt.show()


```
